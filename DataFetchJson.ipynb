```python
def Clean_Data(DataOri):
    Text_to_Num = {'one':1,'two':2,'three':3,'four':4,'five':5}
    Cleaned_Data = []
    unique_users = set()
    
    for user in DataOri:
        row_Data_Rating=str(user["rating"]).strip().lower()
        if(row_Data_Rating in Text_to_Num):
            row_Data_Rating=Text_to_Num[row_Data_Rating]
        user["rating"] = row_Data_Rating
        # hendle missing value
        row_age = user.get("age")
        if(row_age == None):
            user["age"]=None

        #Deduplication
        if(user["name"].strip() is unique_users):
            continue
        unique_users.add(user["rating"])
        Cleaned_Data.append(user)
    return Cleaned_Data
```

```python
Data=Clean_Data(DataOri)
Data
```

```python
def get_insights(data):
    
    if len(data) == 0:
        print("No data available")
        return
    
    tot_rating = 0
    poor_ratings = 0

    for user in data:
        try:
            rating = float(user["rating"])
        except:
            continue   # skip invalid data

        tot_rating += rating

        if rating < 3:
            poor_ratings += 1

    avg = tot_rating / len(data)
    poor_percent = (poor_ratings / len(data)) * 100

    print(f"avg rating = {avg}")
    print(f"% of user with poor rating = {poor_percent}%")
```

```python
get_insights(Data)
```
